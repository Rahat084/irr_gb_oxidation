```python
import mms

fs, ss = mms.evaluate(
    (
        "rho * cp * diff(u,t) - div(k*grad(u)) - "
        "shortwave*sin(0.5*x*pi)*exp(kappa*y)*sin(1/(hours*3600)*pi*t)"
    ),
    "t*sin(pi*x)*sin(5*pi*y)",
    scalars=["rho", "cp", "k", "kappa", "shortwave", "hours"],
)
mms.print_hit(fs, "mms_force")
mms.print_hit(ss, "mms_exact")
```
